```python
# Tree_Traversal
class Tree:
    def __init__(self, i, l, r) :
        self.index = i
        self.left = l
        self.right = r

    def addNode(self, i, l, r) :
        if self.index == None or self.index == i :
            self.index = i
            self.left = Tree(l, None, None) if l != None else None
            self.right = Tree(r, None, None) if r != None else None

            return True

        else :
            flag = False

            if self.left != None :
                flag = self.left.addNode(i, l, r)

            if flag == False and self.right != None :
                flag = self.right.addNode(i, l, r)

            return flag

def preorder(tree) :
    '''
    tree를 전위순회 하여 리스트로 반환
    Root - L - R
    '''
    result = []
    
    result.append(tree.index)
    
    if tree.left != None :
        result = result + preorder(tree.left)   
        #tree의 left를 전위순회한 값을 return해서 result에 add
    if tree.right != None :
        result = result + preorder(tree.right)

    return result

def inorder(tree) :
    '''
    tree를 중위순회 하여 리스트로 반환
    L - Root - R
    '''
    result = []

    if tree.left != None:
        result = result + inorder(tree.left)    
        #left가 있으면 중위순회해서 return
    result.append(tree.index)
    
    if tree.right != None :
        result = result + inorder(tree.right)

    return result

def postorder(tree) :
    '''
    tree를 후위순회 하여 리스트로 반환
    L - R - Root
    '''
    result = []
    
    if tree.left != None:
        result = result + postorder(tree.left)
    if tree.right != None:
        result = result + postorder(tree.right)
    result.append(tree.index)
    
    return result

def main():
    myTree = Tree(None, None, None)

    n = int(input())

    for i in range(n) :
        myList = [int(v) for v in input().split()]

        if myList[1] == -1 :
            myList[1] = None

        if myList[2] == -1 :
            myList[2] = None

        myTree.addNode(myList[0], myList[1], myList[2])

    print(*preorder(myTree))
    print(*inorder(myTree))
    print(*postorder(myTree))


if __name__ == "__main__":
    main()
```

```python
# Tree 구현
class Tree:
    '''
    tree: 코드가 실행되고 있는 상태를 나타내는 자료구조
    function을 어떤 순서로 실행시킬 것인지
    코드 실행 = 트리를 후위순회
    '''
    def __init__(self, i, l, r) :
        self.index = i
        self.left = l
        self.right = r

    def addNode(self, i, l, r) :
        '''
        트리 내의 정점 i에 대하여 왼쪽자식을 l, 오른쪽 자식을 r로 설정해주는 함수
        '''
      
        if self.index == i or self.index == None:
            self.index = i
            
            if l == None : 
                self.left = None
            else: 
                self.left = Tree(l, None, None)
            if r == None:
                self.right = None
            else: 
                self.right = Tree(r, None, None)
                
            return True
                
        else: 
            if self.left != None:
                flag = self.left.addNode(i, l, r)  
                # left subtree에 i 있으면 addnode
                # add or not: flag에 true or false 
                # -> right subtree에 넣을지 말지 결정하기 위해
            else:
                flag = False
            
            if flag == True:
                return True
            else:
                if self.right != None:
                    flag = self.right.addNode(i, l ,r)
                else:
                    flag = False
                    
                if flag == True:
                    return True
                else:
                    return False
                
```

```python
# Lec3_HW1_primenumber
import math

def isPrime(n):
    '''
    숫자 n이 소수이면 True, 아니면 False를 반환하는 함수
    '''
    b = int(math.sqrt(n)+1)
    a = 0
    for i in range(1,b):
        if n % i == 0:
            a += 1
    if a > 1:
        return False
    else:
        return True

def main():
    n = int(input())

    print(isPrime(n))

if __name__ == "__main__":
    main()
```
