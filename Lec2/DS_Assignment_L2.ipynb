```python
class Stack:
    def __init__(self) :
        self.myData = []

    def push(self, n) :
        self.myData.append(n)
        

    def pop(self) :   #stack에서 가장 위에 있는 정수를 제거 / 만약 stack에 아무 원소가 없다면 아무 일도 하지 않기
        if len(self.myData) == 0:
            return
        else:
            self.myData = self.myData[:-1]
            ##self.myData.pop(): 맨뒤에꺼 날려
        

    def size(self) :  #stack에 들어 있는 정수의 개수를 return 합니다.
        return len(self.myData) 

    def empty(self) : #stack이 비어있다면 1, 아니면 0을 return 합니다.
        if len(self.myData) == 0:
            return 1
        else:
            return 0
            

    def top(self) :   #stack의 가장 위에 있는 정수를 return / 만약 stack에 들어있는 값이 없을 경우에는 -1을 return
        if len(self.myData) == 0:
            return -1
        else:
            return self.myData[-1]

def main():
    stack = Stack()
    stack.push(1)
    stack.push(2)
    stack.push(4)
    stack.pop()

    print(stack.size()) 
    print(stack.top())


if __name__ == "__main__":
    main()
```

```python
class Stack:
    def __init__(self) :
        self.myData = []

    def push(self, n) :
        self.myData.append(n)
        

    def pop(self) :
        if len(self.myData) == 0:
            return
        else:
            self.myData = self.myData[:-1]
            ##self.myData.pop(): 맨뒤에꺼 날려
        

    def size(self) :
        return len(self.myData)

    def empty(self) :
        if len(self.myData) == 0:
            return 1
        else:
            return 0
            

    def top(self) :
        if len(self.myData) == 0:
            return -1
        else:
            return self.myData[-1]

def checkParen(p) :   #괄호 문자열 p의 쌍이 맞으면 "YES", 아니면  "NO"를 반환
    myStack = Stack()
    
    for x in p:
        if x == '(':
            myStack.push(x)
        else:
            if myStack.empty() == 1:
                return "NO"
            else:
                myStack.pop()
                
                
    if myStack.empty() == 1:
        return 'YES'
    else:
        return "NO"

def main():
    x = input()
    print(checkParen(x))

if __name__ == "__main__":
    main()
```

```python
class Queue:
    def __init__(self) :
        self.myData = []
        

    def push(self, n) :
        self.myData.append(n)

    def pop(self) :   #queue에서 가장 앞에 있는 정수를 제거, 만약 queue에 들어있는 값이 없을 경우에는 아무 일도 하지 않기
    
        if self.myData:
            self.myData = self.myData[1:len(self.myData)]
        else:
            return

    def size(self) :

        return len(self.myData)

    def empty(self) :
        if self.myData:
            return 0
        else:
            return 1

    def front(self) :
        if self.myData:
            return self.myData[0]
        else:
            return -1

    def back(self) :
        return self.myData[len(self.myData)-1]




def main():
    queue = Queue()
    queue.push(1)
    queue.push(2)
    queue.pop()
    queue.push(4)
    queue.pop()


    print(queue.size()) 
    print(queue.front())
    print(queue.empty())


if __name__ == "__main__":
    main()
```

```python
def findOrder(p) :
    '''
    괄호 p가 주어질 때, 각 괄호가 몇 번째로 계산되어야 하는 괄호인지를 list로 반환합니다.

    예를 들어, p='(()())' 일 경우, [3, 1, 1, 2, 2, 3] 을 반환합니다.
    '''
    n = len(p)
    result = [0]*n
    myStack_left = []
    myStack_right = []
    a = 0
    b = 1
    c = 1
    for x in p:
        if x == '(':
            myStack_left.append(x)
            myStack_left.append(b)
            
            result[a] = b
            b += 1
        else:
            myStack_right.append(x)
            myStack_left.append(c)
            result[a] = c
            c += 1
        a += 1
    return result

def main():
    p = input()
    print(*findOrder(p))

if __name__ == "__main__":
    main()
```
