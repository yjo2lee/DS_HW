```python
class Stack:
    def __init__(self) :
        self.myData = []

    def push(self, n) :
        self.myData.append(n)
        

    def pop(self) :
        if len(self.myData) == 0:
            return
        else:
            self.myData = self.myData[:-1]
            ##self.myData.pop(): 맨뒤에꺼 날려
        

    def size(self) :
        return len(self.myData)

    def empty(self) :
        if len(self.myData) == 0:
            return 1
        else:
            return 0
            

    def top(self) :
        if len(self.myData) == 0:
            return -1
        else:
            return self.myData[-1]

def checkParen(p) :   #괄호 문자열 p의 쌍이 맞으면 "YES", 아니면  "NO"를 반환
    myStack = Stack()
    
    for x in p:
        if x == '(':
            myStack.push(x)
        else:
            if myStack.empty() == 1:
                return "NO"
            else:
                myStack.pop()
                
                
    if myStack.empty() == 1:
        return 'YES'
    else:
        return "NO"

def main():
    x = input()
    print(checkParen(x))

if __name__ == "__main__":
    main()
```

```python
class Queue:
    def __init__(self) :
        self.myData = []
        

    def push(self, n) :
        self.myData.append(n)

    def pop(self) :   #queue에서 가장 앞에 있는 정수를 제거, 만약 queue에 들어있는 값이 없을 경우에는 아무 일도 하지 않기
    
        if self.myData:
            self.myData = self.myData[1:len(self.myData)]
        else:
            return

    def size(self) :
    
        queue에 들어 있는 정수의 개수를 return 합니다.
        '''
        return len(self.myData)

    def empty(self) :
        '''
        queue이 비어있다면 1, 아니면 0을 return 합니다.
        '''
        if self.myData:
            return 0
        else:
            return 1

    def front(self) :
        '''
        queue의 가장 앞에 있는 정수를 return 합니다. 만약 queue에 들어있는 값이 없을 경우에는 -1을 return 합니다.
        '''
        if self.myData:
            return self.myData[0]
        else:
            return -1

    def back(self) :
        '''
        queue의 가장 뒤에 있는 정수를 return 합니다. 만약 queue에 들어있는 값이 없을 경우에는 -1을 return 합니다.
        '''
        return self.myData[len(self.myData)-1]




def main():
    queue = Queue()

    '''
    테스트를 하고싶으면, 아래 부분을 수정합니다.
    '''
    queue.push(1)
    queue.push(2)
    queue.pop()
    queue.push(4)
    queue.pop()


    print(queue.size()) 
    print(queue.front())
    print(queue.empty())


if __name__ == "__main__":
    main()
```
