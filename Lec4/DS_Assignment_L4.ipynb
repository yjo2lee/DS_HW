```python
# Lec4_Q1_Heap_PriorityQue
class priorityQueue:

    def __init__(self) :
        self.data = [0]

    def push(self, value) :
        '''
        우선순위 큐에 value를 삽입
        '''
        self.data.append(value)
        index = len(self.data)-1    #나의 위치
        
        while index != 1:   #root되면 stop
            if self.data[index//2] > self.data[index]:
                #move to the top
                temp = self.data[index//2]
                self.data[index//2] = self.data[index]
                self.data[index] = temp
                
                index = index // 2

            else:   #priority맞으면 break
                break

    def top(self) :
        '''
        우선순위가 가장 높은 원소를 반환. 만약 우선순위 큐가 비어있다면 -1을 반환.
        '''
        if len(self.data) == 1:
            return -1
        else: 
            return self.data[1]


    def pop(self) :
        '''
        우선순위가 가장 높은 원소를 제거.
        '''
        if len(self.data) == 1:
            return
            
        self.data[1] = self.data[-1]
        self.data.pop()     #remove the last element 
        
        index = 1
        
        while True:
            priority = -1  #compare bw left and right -> priority_indicate
            # 1 no children
            # 2 only left child
            # 3 both children
        
            #number of elements = 9 // index = 8
            if len(self.data)-1 < index*2: #1
                break
            
            elif len(self.data)-1 < index*2+1: #2
                priority = index*2
            
            else:
                if self.data[index*2] < self.data[index*2+1]:
                    priority = index*2
                else:
                    priority = index*2+1
                
            ####
            # compare between index and priority
            if self.data[index] > self.data[priority]:
                temp = self.data[priority]
                self.data[priority] = self.data[index]
                self.data[index] = temp
                index = priority
                
            else:
                break
                
# 1. 우선순위 큐를 이용하면 할 수 있는 것들
#    - 정렬 (여러개의 숫자가 주어질 때 이를 정렬하라)
#    - push all + pop all
# 2. 구글링하면 다 나와요

def main():
    myPQ = priorityQueue()

    myPQ.push(1)
    myPQ.push(4)
    myPQ.push(3)
    myPQ.push(2)

    print(myPQ.top())
    myPQ.pop()

    print(myPQ.top())
    myPQ.pop()

if __name__ == "__main__":
    main()
```

```python
# Lec4_Q2_Recursive function example

LIMIT_NUMBER = 1000000007

def getPower(m, n):
    '''
    m^n 을 LIMIT_NUMBER로 나눈 나머지를 반환하는 함수를 작성
    '''
    if n == 0:  #initial value
        return 1
    
    if n % 2 == 0:
        temp = getPower(m, n//2)
        return (temp * temp) % LIMIT_NUMBER
    else:
        return getPower(m,n-1) * m % LIMIT_NUMBER
    

def main():
    myList = [int(v) for v in input().split()]

    print(getPower(myList[0], myList[1]))

if __name__ == "__main__":
    main()
    
# 재귀호출: function의 기능은 된다고 생각하고(funciton의 정의가 중요)
# (n-1), (n) 과의 link가 중요
```
